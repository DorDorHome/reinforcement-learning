```python
print("Action space size: {}".format(env.action_space.n))
print(env.get_action_meanings())  # env.unwrapped.get_action_meanings() for gym 0.8.0 or later

observation = env.reset()
print("Observation space (first component) shape: {}".format(observation[0].shape))

# Loop through each step and render the environment
for step in range(10):
    plt.figure()
    

    plt.imshow(env.render())  # Add mode='rgb_array' if necessary
    
    plt.title(f"Step: {step}")
    plt.axis('off')  # Hide the axis to only show the game screen
    plt.show()  # Display the plot
    
    # Perform the action (2 in this case) and get the new observation, reward, done, info
    env.step(3)
    
    

env.close()
```
